## Chicago Divvy Bikeshare program information



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
font = {'weight' : 'bold',
        'size'   : 22}

plt.rc('font', **font)


df = pd.read_csv('/content/chicagoDivvy.csv')
df.info();
# here we have imported and aliased the necesarry libraries and also pre set the size of the fonts while reading and labeling the csv to df
```

What is the percentage of men and women using the rideshare program?

```python
df.gender.value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(13,10)); # here we take the valut_counts() which gives us unique values and ploted them using a pie chart
# autopct is helping us with labelling the percentages and figsize lets you set the size of the chart making it easier to be read. 
```

Which days see the most use of the bicycle share program?



```python
week = df.iloc[:, [4,5]] # here we are designating week to be the alias to just pull the days and hours columns using iloc
week.head(5)
```

```python
weekplt = week.groupby('day').sum() # next we assign to weekplt the data grouped by days and summing the hours in them. 
```

What are the 5 most and least visited station names and where are they located?

```python
station = df.loc[:,['from_station_name','latitude_start', 'longitude_start']] # creating station from original df and pinpointing the three columns
station
```

```python
stationTop = station.value_counts().head(5).to_frame().reset_index() 
# counting the values that are in the table and getting the top 5 then .to_frame to put back into df instead of series and resetting index
stationBot = station.value_counts().tail(5).to_frame().reset_index() 
# counting the values that are in the table and getting the bottom 5 then .to_frame to put back into df instead of series and resetting index
```

```python
stationTop
```

```python
import folium # folium will help run the map
```

```python
# setting up the map with the biginning zoom and location for chicago
myMap = folium.Map(
    location=[41.88183246, -87.62317685],
    zoom_start=11
)
```

```python
# for loop to add each 5 markers giving them different colors and adding the hovber and view for the names on the markers
for _, markerTop in stationTop.iterrows():
  folium.Marker(
      location=[markerTop['latitude_start'], markerTop['longitude_start']],
      popup=markerTop['from_station_name'],
      tooltip=markerTop['from_station_name'],
      icon=folium.Icon(color='green')
  ).add_to(myMap)

for _, markerBot in stationBot.iterrows():
  folium.Marker(
      location=[markerBot['latitude_start'], markerBot['longitude_start']],
      popup=markerBot['from_station_name'],
      tooltip=markerBot['from_station_name'],
      icon=folium.Icon(color='red')      
  ).add_to(myMap)
myMap
```
